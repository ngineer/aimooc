## Read the temperature data

```python
df = pd.read_csv("/content/drive/My Drive/AIMOOC/Modules/Module4/Data/temperatures.csv")
df 
```

## Count countries and months that experienced cooling or warming in 1961, 1990, and 2019

```python
print('In 1961, how many countries and months experienced:')
print("warming: ", df[df["Y1961"]>=0].count()["Y1961"])
print("cooling: ", df[df["Y1961"]<0].count()["Y1961"])
print('In 1990, how many countries and months experienced:')
print("warming: ", df[df["Y1990"]>=0].count()["Y1990"])
print("cooling: ", df[df["Y1990"]<0].count()["Y1990"])
print('In 2019, how many countries and months experienced:')
print("warming: ", df[df["Y2019"]>=0].count()["Y2019"])
print("cooling: ", df[df["Y2019"]<0].count()["Y2019"]) 
```

## Histogram of the temperature difference over 3 years: 1961, 1990, and 2019

```python
plt.hist(df["Y1961"], bins=20, alpha=0.3, color="blue")
plt.hist(df["Y1990"], bins=20, alpha=0.3, color="orange")
plt.hist(df["Y2019"], bins=20, alpha=0.3, color="green")
plt.xlabel("Bins (temperature difference value)")
plt.ylabel("Count")
plt.title("How Many Countries and Months Fall in each Bin") 
```

## Look at the world temperature difference (not just one country)

```python
world = df.loc[(df['Area'] == "World")]
world 
```

## Plot the world temperature difference over all years

```python
worldavg = world.mean()
plt.figure(figsize=(15,5))
plt.xticks(rotation=90)
plt.plot(worldavg)
print(worldavg.max()-worldavg.min())
plt.xlabel("Year")
plt.ylabel("Temperature difference to baseline")
plt.title("World Temperature Difference (averaged over 12 months) over Many Years") 
```
