```python
import pandas as pd
import ipywidgets as widgets
import numpy as np
import matplotlib.pyplot as plt
import sklearn.linear_model as lm 
```

```python
df = pd.read_csv("/content/drive/My Drive/AIMOOC/Modules/Module4/Data/marketing.csv")
df
```

```python
def plot_graph(income=0, amt_spent=0):
    x_values = np.linspace(0, 160000, 160000)
    # run linear regression on Income vs. Meat Purchase information and plot it
    x = df[["Income"]]
    y = df["MntMeatProducts"]
    lin_reg = lm.LinearRegression()
    lin_reg.fit(x, y) 
    plt.scatter(x, y)
    ypred = lin_reg.predict(x)
    plt.plot(x, ypred, color="purple")

    # run linear regression on Income vs. Fish Purchase information and plot it
    x = df[["Income"]]
    y = df["MntFishProducts"]
    lin_reg = lm.LinearRegression()
    lin_reg.fit(x, y) 
    plt.scatter(x, y)
    ypred = lin_reg.predict(x)
    plt.plot(x, ypred, color="red")

    # add plot labels
    plt.xlabel("Income")
    plt.ylabel("Amount spent on meat or fish")
    plt.title("Best fit lines for customers income vs. spending habits")
    plt.ylim([0,1000])
    plt.plot(x_values, 0*x_values + amt_spent, color='black')
    plt.axvline(x=income, color='black')

widgets.interact(plot_graph, income=(0,160000,5000), amt_spent=(10, 1000, 10) )
```

## Dataset

The code cells above load a csv file containing information about the spending habits of customers based on their income. 

Run all the cells in order to produce a plot of Income on the x-axis, and amount spent on meat (blue dots) or fish (orange dots) on the y-axis. 

The plot also includes the result of running linear regression on the Income vs. Meat Spending (purple line) and Income vs. Fish spending (red line). 

The plot is interactive. 
* You can move the slider labeled `income` to move a black vertical line to different income values. 
* You can move the slider labeled `amt_spent` to move a black horizontal line to different amount spent values.
