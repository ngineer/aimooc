Label = 0: does not spend a lot on meat products

Label = 1: spends a lot on meat products

Use a support vector linear classifier to classify datapoints with x = Income and y = MntWines as being likely to spend a lot on meat products (label = 0) or being unlikely to spend a lot on meat products (label = 1).

In addition, create a scatter plot of the data coloring each datapoint with label 0 as red and label 1 as blue.

```python
#Support Vector Linear Classifier
import sklearn.svm as svm

x = df[["Income", "MntWines"]]
y = df["Label"] 

clf = svm.SVC(kernel='linear')
clf.fit(x, y) 
score = clf.score(x, y)
print(score)

def draw_svc_line():
  w = clf.coef_[0]
  slope = -w[0] / w[1]
  b = -clf.intercept_[0] / w[1]
  col1 = x.iloc[:, 0] # get the first column of x
  col2 = x.iloc[:, 1] # get the second column of x
  xx = range(int(col1.min()), int(col1.max()))
  yy = slope * xx + b
  plt.ylim((int(col2.min()), int(col2.max())))
  line = plt.plot(xx, yy, color='orange')
  plt.scatter(df[df["Label"] == 0]["Income"],df[df["Label"] == 0]["MntWines"], color = "red")
  plt.scatter(df[df["Label"] == 1]["Income"],df[df["Label"] == 1]["MntWines"], color = "blue")
  plt.xlabel("Income")
  plt.ylabel("Wine")
  plt.title("Income vs. Wine")

draw_svc_line()  
```
