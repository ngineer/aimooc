# Part 3

We provide you with the following function to model a set of dependent variables (x_columns) to a single independent variable (y_column) using linear regression. 

For example, you can relate Shoe Size and Days to Sale to Sale Price using the following:

do_ML_fit(df[["Shoe Size","Days to Sale"]], df["Sale Price"])

```python
def do_ML_fit(x_columns, y_column):
  lin_reg = lm.LinearRegression()
  lin_reg.fit(x_columns, y_columns) 
  print('Coefficients:', lin_reg.coef_)
  print('Intercept:', lin_reg.intercept_)
  print(lin_reg.score(x_columns, y_column))
  print()
  return lin_reg
```
