```python
import matplotlib.pyplot as plt
import sklearn.linear_model as lm 
import pandas as pd
```

```python
import warnings
from sklearn.exceptions import DataConversionWarning
warnings.filterwarnings(action='ignore')
```

```python
df = pd.read_csv("/content/drive/My Drive/AIMOOC/Modules/Module2/Data/avg_h_w.csv")
df
```

```python
x = df[["Height"]]
y = df["Weight"]
lin_reg = lm.LinearRegression()
lin_reg.fit(x, y) 
print('Coefficients:', lin_reg.coef_)
print('Intercept:', lin_reg.intercept_)

plt.scatter(x, y)
ypred = lin_reg.predict(x)
plt.plot(x, ypred, color="orange")
plt.xlabel("Height (Meters)")
plt.ylabel("Weight (Kilograms)")
plt.title("Linear relationship between Height and Weight")
```

```python
pred_weight = lin_reg.predict([[1.61]])
print(pred_weight)
```

```python
pred_weight = lin_reg.predict([[1.9]])
print(pred_weight)
```

## Water Dataset

```python
df_water = pd.read_csv("/content/drive/My Drive/AIMOOC/Modules/Module2/Data/water_potability.csv")
df_water = df_water.dropna()
df_water
```

```python
x = df_water[["ph"]]
y = df_water["Hardness"]
lin_reg = lm.LinearRegression()
lin_reg.fit(x, y) 
print('Coefficients:', lin_reg.coef_)
print('Intercept:', lin_reg.intercept_)
print('Score:', lin_reg.score(x,y))

plt.scatter(x, y)
ypred = lin_reg.predict(x)
plt.plot(x, ypred, color="orange")
plt.xlabel("ph Level")
plt.ylabel("Hardness")
plt.title("Linear relationship between ph and hardness water levels")
```

```python
pred_hardness = lin_reg.predict([[7]])
print(pred_hardness)
```
