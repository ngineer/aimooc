### Reading Data
Read in the data file into a DataFrame where one column is the gallons of gas used and the second column is the miles traveled on that amount of gas for a Toyota Corolla. 

```
import pandas as pd
df = pd.read_csv("/content/drive/My Drive/AIMOOC/Modules/Module2/Data/car_toyota_corolla.csv")
df
```

Recall that when we manually tried to fit a line to this data by picking arbitrary sets of 2 points, we got all sorts of lines. But the line that best represents all the data is the one that minimizes the mean-squared-error. 

### Finding the Best Fit Line
Use linear regression to find the formula for the best fit line.


```
import sklearn.linear_model as lm 
```

```
x = df[["Gallons"]]
y = df["Miles"]
lin_reg = lm.LinearRegression()
lin_reg.fit(x, y) 
print('Coefficients:', lin_reg.coef_)
print('Intercept:', lin_reg.intercept_)
df
```

```
import matplotlib.pyplot as plt
plt.scatter(x, y)
#ypred = lin_reg.predict(x)
#plt.plot(x, ypred, color="blue")
plt.xlabel("Gallons")
plt.ylabel("Miles")
plt.title("How Many Miles Toyota Corolla Travels on Some Number of Gallons")
```

```
plt.scatter(x, y)
y2 = lin_reg.coef_ * x + lin_reg.intercept_
plt.plot(x, y2, color="orange")

plt.xlabel("Gallons")
plt.ylabel("Miles")
plt.title("How Many Miles Toyota Corolla Travels on Some Number of Gallons")
```

## Multicar Dataset

```
df2 = pd.read_csv("/content/drive/My Drive/AIMOOC/Modules/Module2/Data/multicar.csv")
df2
```

```
plt.scatter(df2["Gallons"], df2["Miles"])
plt.xlabel("Gallons")
plt.ylabel("Miles")
plt.title("How Many Miles a Car Runs on Some Number of Gallons (for 8 Cars)")
```

```
x2 = df2[["Gallons"]]
y2 = df2["Miles"]
lin_reg2 = lm.LinearRegression()
lin_reg2.fit(x2, y2) 
print('Coefficients:', lin_reg2.coef_)
print('Intercept:', lin_reg2.intercept_)
```

```
plt.scatter(x2, y2)
ypred2 = lin_reg2.predict(x2)
plt.plot(x2, ypred2, color="blue")
plt.xlabel("Gallons")
plt.ylabel("Miles")
plt.title("How Many Miles 8 Different Cars Run (and Best Fit Line)")
```
