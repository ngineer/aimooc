```
import pandas as pd
import matplotlib.pyplot as plt
```

## Read Data
The data contains 2 columns
* bus wait times between 0 and 60 mins
* height data in ft

```
pd.set_option('display.max_columns', 500)
df = pd.read_csv("/content/drive/My Drive/AIMOOC/Modules/Module3/Data/distributions.csv")
df
```

## Plot the data

```
plt.figure()
plt.scatter(df.index,df['Uniform'])
plt.xlabel('Day')
plt.ylabel('Time for Bus to Arrive (min)')
```

```
plt.figure()
plt.scatter(df.index,df['Gaussian'])
plt.xlabel('Person')
plt.ylabel('Height (ft)')
```

## Mean and Standard Deviation for Uniform Data

```
print("Uniform:")
print(df["Uniform"].mean())
print(df["Uniform"].std())
```

## Mean and Standard Deviation for Gaussian Data

```
print("Gaussian")
print(df["Gaussian"].mean())
print(df["Gaussian"].std())
```
